```python
import numpy as np
import matplotlib.pyplot as plt

#Q1. Ask the user to enter a value (e.g., M) for a mathematical function.
# Generate x values from -10 to 10 using np.linspace().
# Compute y values for:
# y= Mx
#2
#y=Msin(x)
# Plot both functions in a single figure:
# Use different colors and line styles.
# Add legend, grid, and title.


M=float(input("Enter a value for M: "))
x=np.linspace(-10,10,100)
y1=M*x**2
y2=M*np.sin(x)


plt.figure(figsize=(8,6))
plt.plot(x,y1,label="y=Mx",color="b",linestyle="-")
plt.plot(x,y2,label="y=Msin(x)",color="r",linestyle="--")


plt.xlabel("x values")
plt.ylabel("y values")
plt.title("Graphs of y=Mx and y=Msin(x)")
plt.legend()
plt.grid(True)


plt.show()
```

```python
#Q.2 Create a dataset of five subjects and scores.
#Convert it into a Pandas DataFrame.
# Plot the scores using a Seaborn bar plot with:
# Different colors for each bar.
# Annotations on top of each bar.
# Title, axis labels, and grid.
import seaborn as sns
import pandas as pd
data={
    'Subject':['physics','maths','chemistry','english','sociology'],
    'Score':[95,88,75,96,90]
}
print(data)

df=pd.DataFrame(data)
print(df)

sns.set_style('darkgrid')

ax=sns.barplot(x='Subject',y='Score',data=df,palette='viridis',hue='Subject',legend=False)

for index, row in df.iterrows():
    ax.text(index, row.Score + 1, str(row.Score), ha='center')

plt.title('Subjects vs Scores')
plt.xlabel('Subjects')
plt.ylabel('Scores')
plt.grid(axis='y')
plt.show()

```

```python
#Q3. Write a Python script to select your roll number as seed for NumPy and then generate a dataset of 50 values using np.random.randn().
#Create a 2x2 subplot layout displaying:
# Line plot showing cumulative sum.
# Scatter plot with random noise.
# Customize with titles, axis labels, and grids.

import numpy as np
import matplotlib.pyplot as plt


np.random.seed(102317097)


data=np.random.randn(50)


fig,ax=plt.subplots(2,2,figsize=(10,8))


ax[0,0].plot(np.cumsum(data),color="b")
ax[0,0].set_title("Cumulative Sum")
ax[0,0].set_xlabel("Index")
ax[0,0].set_ylabel("Cumulative Value")
ax[0,0].grid(True)

ax[0,1].scatter(range(50),data,color="r")
ax[0,1].set_title("Random Noise Scatter Plot")
ax[0,1].set_xlabel("Index")
ax[0,1].set_ylabel("Value")
ax[0,1].grid(True)

plt.tight_layout()
plt.show()

```
