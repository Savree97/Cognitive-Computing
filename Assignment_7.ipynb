```python
#Q1. You are a data analyst at an e-commerce company. Your task is to analyze sales data, clean and process it, perform statistical analysis, and generate visual insights to help the business make informed decisions.
#1. Use your roll number as the random seed to generate unique data for each student.
#2. Follow the questions step by step, as each part builds upon the previous one.
#3. Submit the assignment as a Jupyter Notebook (.ipynb) with:
# Well-documented code and clear comments.
# Markdown explanaƟons for each step.
# Screenshots of the plots and outputs.

import numpy as np
import pandas as pd



#Part I: Randomized Sales Data GeneraƟon (NumPy)
#1. Initialize Random Seed

np.random.seed(102317097)

#2. Generate Sales Data
# Create a NumPy array of random sales data for 12 months (shape = (12, 4)),
#where each column represents a product category:
# Electronics, Clothing, Home & Kitchen, Sports.
# Sales values should be between 1000 and 5000 units per month.
# The range should be randomized using the seed.

sales_data = np.random.randint(1000, 5000, size=(12, 4))
months=["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
categories=["Electronics", "Clothing", "Home & Kitchen", "Sports"]

# 3. Convert to Pandas DataFrame

df = pd.DataFrame(sales_data, columns=categories, index=months)
df.head()

```

```python
#Part II: Data Manipulation and Analysis (Pandas)
#1. Display the first 5 rows and summary statistics of the DataFrame.
print(df.head())
print(df.describe())

#2. Calculate the total sales per category and total sales per month.
total_sales_category = df.sum()
total_sales_month = df.sum(axis=1)

print("\nTotal Sales Per Category:\n", total_sales_category)
print("\nTotal Sales Per Month:\n", total_sales_month)

#3. Calculate the average sales growth between consecuƟve months for each category.
df_growth = df.pct_change() * 100
df_growth = df_growth.round(2)
print("\nAverage Sales Growth (%):\n", df_growth)

#4. Add the following columns to the DataFrame:
# Total Sales: Sum of all categories per month.
# Growth Rate: Percentage change in Total Sales from the previous month.
df["Total Sales"] = df.sum(axis=1)
df["Growth Rate"] = df["Total Sales"].pct_change() * 100
df["Growth Rate"] = df["Growth Rate"].round(2)
print("\nUpdated DataFrame:\n", df)

#5. Apply a randomized discount:
# If your roll number is even, apply a 10% discount to the Electronics category.
# If odd, apply a 15% discount to the Clothing category.
df["Clothing"] = df["Clothing"] * 0.85
print("\nDataFrame after Discount:\n", df)
```

```python
#Part III: Visualizations (Matplotlib and Seaborn)
#1. Plot monthly sales trends for each category using line plots.
#2. Create the following plots:
# Box plots to show the sales distribution for each category.

import matplotlib.pyplot as plt
import seaborn as sns

# 1. Line Plot - Monthly Sales Trends
plt.figure(figsize=(10, 6))
for col in categories:
    plt.plot(df.index, df[col], label=col)

plt.title("Monthly Sales Trends")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.legend()
plt.grid(True)
plt.show()

# 2. Box Plots - Sales Distribution for Each Category
plt.figure(figsize=(8,6))
sns.boxplot(data=df[categories])
plt.title("Sales Distribution by Category")
plt.xlabel("Product Categories")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
#Q.2 For the array: array = np.array([[1, -2, 3],[-4, 5, -6]])
 #i. Find element-wise absolute value
import numpy as np

array = np.array([[1, -2, 3], [-4, 5, -6]])
absolute_values = np.abs(array)
print("\nElement-wise Absolute Values:\n", absolute_values)

#ii. Find the 25th, 50th, and 75th percenƟle of flattened array, for each column, for each row.

percentile_25=np.percentile(array, 25, axis=None)
percentile_50=np.percentile(array, 50, axis=None)
percentile_75=np.percentile(array, 75, axis=None)

percentile_col=np.percentile(array, [25, 50, 75], axis=0)
percentile_row=np.percentile(array, [25, 50, 75], axis=1)

print("\nPercentiles (Flattened):", percentile_25, percentile_50, percentile_75)
print("\nPercentiles (Column-wise):\n", percentile_col)
print("\nPercentiles (Row-wise):\n", percentile_row)

#iii. Mean, Median and Standard Deviagion of flattened array, of each column, and each row
mean_all=np.mean(array)
median_all=np.median(array)
std_all=np.std(array)

mean_col=np.mean(array, axis=0)
median_col=np.median(array, axis=0)
std_col=np.std(array, axis=0)

mean_row=np.mean(array, axis=1)
median_row=np.median(array, axis=1)
std_row=np.std(array, axis=1)

print("\nMean, Median, Std (Flattened):", mean_all, median_all, std_all)
print("\nColumn-wise:\n", mean_col, median_col, std_col)
print("\nRow-wise:\n", mean_row, median_row, std_row)
```

```python
#Q.3 For the array: a = np.array([-1.8, -1.6, -0.5, 0.5,1.6, 1.8, 3.0]).
 #i. Find floor, ceiling and truncated value, rounded values

a=np.array([-1.8,-1.6,-0.5,0.5,1.6,1.8,3.0])


floor_values=np.floor(a)
ceiling_values=np.ceil(a)
truncated_values=np.trunc(a)
rounded_values=np.round(a)

print("\nFloor Values:",floor_values)
print("Ceiling Values:",ceiling_values)
print("Truncated Values:",truncated_values)
print("Rounded Values:",rounded_values)
```

```python
#Q.4 WAP to swap two elements in a list in Python Using a Temporary Variable.


def swap_list_elements(lst,index1,index2):
    temp=lst[index1]
    lst[index1]=lst[index2]
    lst[index2]=temp
    return lst

my_list=[2,4,6,8,10]
print("Original List:",my_list)
swapped_list=swap_list_elements(my_list,1,3)
print("Swapped List:",swapped_list)
```

```python
#Q.5 Write a Python program to swap two elements in a set by converƟng it to a list first.

def swap(lst,i,j):
    lst[i],lst[j]=lst[j],lst[i]
    return lst

def swap2(s,elem1,elem2):
    l=list(s)
    i,j=l.index(elem1),l.index(elem2)
    return set(swap(l,i,j))

s={1,2,3,4}
print("Original Set:",s)
s=swap2(s,1,3)
print("Swapped Set:",s)
```
